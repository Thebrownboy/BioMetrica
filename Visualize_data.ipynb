```python
import pandas as pd 
import numpy as np 
from matplotlib import pyplot  as plt 
import seaborn as sns 
```

```python
df = pd.read_csv("dataset/body_level_classification_train.csv")
```

```python
df.columns
```

```python
ax=df["Body_Level"].value_counts().plot(kind="bar",title="Counts of Observation in each class")
plt.ylabel("number of observation")
plt.show()
```

```python
# ax =df["Meal_Count"].plot.box(figsize=(15,15))
ax2=df["Phys_Act"].plot(kind='box',figsize=(15,15))

print(type(ax2))
plt.show()
```

```python
ax1=df["Meal_Count"].plot(kind= "kde",figsize=(15,15))
plt.show()
```

```python
for i in df.columns : 
    if df[i].dtype.name == 'object':
        fig , (ax1,ax2) = plt.subplots(1,2,figsize=(15,10))
        df[i].value_counts().plot(kind="pie",title=f"{i} plot",ax=ax1)
        df[i].value_counts().plot(kind='bar',title=f"{i} plot bars",ax=ax2)
        
    else :
        fig , (ax1,ax2) = plt.subplots(1,2,figsize=(15,10))
        df[i].plot(kind='box',ax=ax1,title =f"{i} Box Plot")
        sns.histplot(df[i],ax=ax2,kde=True)
    plt.show()
    fig.savefig(f"{i}.png")
    




```
